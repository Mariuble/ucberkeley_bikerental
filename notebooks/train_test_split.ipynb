In this notebook, we will prepare the shell scripts to train and test the predictive model.

```python
import pandas as pd
import random
import numpy as np

seed=99
random.seed(seed)
np.random.seed(seed)
```

```python
rental = pd.read_csv('../data/rental.csv')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(rental, test_size=0.2)

X_train.head()
```

```python
X_train.to_csv('../data/train.csv', index=False)
X_test.to_csv('../data/test.csv', index=False)
```
